```python
import numpy as np

from fifth_harmony import *
```

```python
k = 7
omega = 1 # frecuency

y0_vals = [1, 1.5, 2, 2.5, 3] # initial position
vy0_vals = [1, 1.5, 2, 2.5, 3] # initial velocity

# shift to (velocity, equilibrium position)
bx = 0.0
by_vals = np.round(np.arange(0.0, 1.1, 0.1), 1)
```

```python
for y0 in y0_vals:
    for vy0 in vy0_vals:
        for by in by_vals:
            print(f"k={k}, omega={omega}, y0={y0}, vy0={vy0}, bx={bx}, by={by}")
            if data_exists(k, omega, y0, vy0, bx, by):
                print("Data exists, skipping computation!\n")
                continue
            else:
                time_range, data_expected, data_actual = harmonic_oscillator(k, omega, y0, vy0, bx, by, msg=False)
                save_data(k, omega, y0, vy0, bx, by, time_range, data_expected, data_actual)
```
