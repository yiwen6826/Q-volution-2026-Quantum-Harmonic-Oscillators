This Jupyter Notebook showcases how to run many cases of the quantum algorithm in a few lines of code and save the data accordingly inside of a directory named `data`.

Suppose that we would like to solve for `y(t)` and `y'(t)` given the values below.

```python
k = 7
omega = 1 # frecuency

y0_vals = [1, 1.5] # initial position
vy0_vals = [2.5, 3] # initial velocity

# shift to (velocity, equilibrium position)
b0 = 0.0
b1 = 0.6
```

Using the custom module `fifth_harmony` all that needs to be done is use `harmonic_oscillator` inside of some `for` loops. The `data_exists` function checks whether the case we are interested in already exists in the `data` directory.

```python
for y0 in y0_vals:
    for vy0 in vy0_vals:
        print(f"k={k}, omega={omega}, y0={y0}, vy0={vy0}, b0={b0}, b1={b1}")
        if data_exists(k, omega, y0, vy0, b0, b1):
            print("Data exists, skipping computation!\n")
            continue
        else:
            time_range, data_expected, data_actual = harmonic_oscillator(k, omega, y0, vy0, b0, b1, msg=False)
            save_data(k, omega, y0, vy0, b0, b1, time_range, data_expected, data_actual)
```

**Keep in mind**. The directory `data` is automatically created when running `harmonic_oscillator` for the first time. All the files will be saved inside a subdirectory using the naming convention `k_oemga_y0_vy0_b0_b1`.

In case we want to load existing data, the following lines of code will be needed

```python
time, dydt_expected, dydt_actual = load_data(k, omega, y0, vy0, b0, b1, data='dydt')
# possible values of `data` are 'y', 'dydt', 'KE', 'PE'
```
